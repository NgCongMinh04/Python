# ## Thuộc tính của Dataset: Bank Loan Dataset

"""
| Thuộc tính                      | Ý nghĩa                                                                                   |
|---------------------------------|--------------------------------------------------------------------------------------------|
| Loan ID                         | Mã số khoản vay của khách hàng                                                            |
| Customer ID                     | Mã định danh khách hàng                                                                   |
| Loan Status                     | Trạng thái khoản vay (Được duyệt hoặc bị từ chối)                                         |
| Current Loan Amount             | Số tiền khoản vay hiện tại                                                                |
| Term                            | Kỳ hạn vay (ngắn hạn hoặc dài hạn)                                                        |
| Credit Score                    | Điểm tín dụng của khách hàng (đánh giá mức độ uy tín tài chính)                          |
| Annual Income                   | Thu nhập hàng năm của khách hàng                                                          |
| Years in current job            | Số năm làm việc tại công việc hiện tại                                                    |
| Home Ownership                  | Hình thức sở hữu nhà ở (Thuê, sở hữu, thế chấp…)                                          |
| Purpose                         | Mục đích vay vốn (mua nhà, mua xe, kinh doanh, tiêu dùng…)                                |
| Monthly Debt                    | Khoản nợ phải trả hàng tháng của khách hàng                                               |
| Years of Credit History         | Số năm có lịch sử tín dụng (thời gian có phát sinh vay mượn ngân hàng)                   |
| Months since last delinquent    | Số tháng kể từ lần trễ hạn thanh toán gần nhất                                            |
| Number of Open Accounts         | Số lượng tài khoản tín dụng đang hoạt động                                                |
| Number of Credit Problems       | Số lượng vấn đề tín dụng đã phát sinh (như nợ xấu, trễ hạn...)                           |
| Current Credit Balance          | Số dư nợ tín dụng hiện tại                                                               |
| Maximum Open Credit             | Hạn mức tín dụng tối đa có thể vay                                                        |
| Bankruptcies                    | Số lần phá sản nếu có                                                                     |
| Tax Liens                       | Số lượng khoản thuế chưa trả hoặc bị phong tỏa tài sản                                    |
"""


```python
# Tiền Xử Lý Dữ Liệu 
# Copy 1 file dữ liệu qua 1 biến 
df = data.copy()
```

```python
df.drop(['Loan ID', 'Customer ID'], axis=1, inplace=True)
```

```python
# Điền trung bình cho cột số liệu quan trọng
df['Credit Score'].fillna(df['Credit Score'].mean(), inplace=True)
df['Annual Income'].fillna(df['Annual Income'].mean(), inplace=True)
```

```python
# Điền mode (giá trị phổ biến nhất) cho cột phân loại
df['Years in current job'].fillna(df['Years in current job'].mode()[0], inplace=True)
```

```python
# Xóa cột thiếu quá nhiều dữ liệu
df.drop('Months since last delinquent', axis=1, inplace=True)
```

```python
# Điền giá trị trung bình cho các cột số còn lại bị thiếu ít 
cols_fill_mean = ['Maximum Open Credit', 'Bankruptcies', 'Tax Liens']

for col in cols_fill_mean:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
# Kiểm tra xem còn lỗi không 
missing_data = df.isnull().sum()
print(missing_data[missing_data > 0])
print("\nSố Lượng Thuộc Tính bị lỗi",missing_data[missing_data > 0].count())
```

```python
df.dropna(inplace=True)
```

```python
missing_data = df.isnull().sum()
print(missing_data[missing_data > 0])
print("\nSố Lượng Thuộc Tính bị lỗi",missing_data[missing_data > 0].count())
```

```python
# Hoàn tất xử lý dữ liệu lỗi 
```

```python
df.head()
```

```python
# Xử lý biểu diễn dữ liệu cho dễ phân tích 
# xử lý Years in current job thành dạng số 
# Hàm xử lý chuyển đổi dữ liệu
def convert_years(x):
    if pd.isnull(x):
        return 0  # nếu thiếu dữ liệu thì gán tạm là 0
    if '<' in x:
        return 0  # nhỏ hơn 1 năm = 0 năm
    if '10+' in x:
        return 10  # trên 10 năm = 10 năm
    # Trích số năm trong chuỗi
    return int(x.split()[0])
# Áp dụng hàm xử lý vào cột Years in current job
df['Years in current job'] = df['Years in current job'].apply(convert_years)

# Kiểm tra lại kết quả
print(df['Years in current job'].unique())
print(df.info())
```

```python
# Làm tròn cột 'Credit Score' còn 2 chữ số thập phân
df['Credit Score'] = df['Credit Score'].round(2)
# Kiểm tra lại dữ liệu sau khi làm tròn
df.head()
```

```python
# Đổi tên cột 
df.rename(columns={
    'Loan Status': 'Trạng thái vay',
    'Current Loan Amount': 'Số tiền vay hiện tại',
    'Term': 'Kỳ hạn vay',
    'Credit Score': 'Điểm tín dụng',
    'Annual Income': 'Thu nhập năm',
    'Years in current job': 'Số năm làm việc',
    'Home Ownership': 'Hình thức nhà ở',
    'Purpose': 'Mục đích vay',
    'Monthly Debt': 'Nợ hàng tháng',
    'Years of Credit History': 'Số năm lịch sử tín dụng',
    'Number of Open Accounts': 'Số tài khoản đang hoạt động',
    'Number of Credit Problems': 'Số vấn đề tín dụng',
    'Current Credit Balance': 'Số dư tín dụng hiện tại',
    'Maximum Open Credit': 'Hạn mức tín dụng tối đa',
    'Bankruptcies': 'Số lần phá sản',
    'Tax Liens': 'Số lần nợ thuế'
}, inplace=True)
```

```python
df.head()
```

```python
# đổi dữ liệu các cột 
# Đổi giá trị trong cột 'Loan Status'
df['Trạng thái vay'] = df['Trạng thái vay'].replace({
    'Fully Paid': 'Đã thanh toán',
    'Charged Off': 'Không thanh toán'
})

# Đổi giá trị trong cột 'Kỳ hạn vay'
df['Kỳ hạn vay'] = df['Kỳ hạn vay'].replace({
    'Short Term': 'Ngắn hạn',
    'Long Term': 'Dài hạn'
})

# Đổi giá trị trong cột 'Hình thức nhà ở'
df['Hình thức nhà ở'] = df['Hình thức nhà ở'].replace({
    'Rent': 'Thuê',
    'Own': 'Sở hữu',
    'Mortgage': 'Thế chấp',
    'Other': 'Khác', 
    'Own Home': 'Nhà riêng',
    'Home Mortgage': 'Thế chấp nhà',
    'HaveMortgage': 'Có thế chấp'
})

# Đổi giá trị trong cột 'Mục đích vay'
df['Mục đích vay'] = df['Mục đích vay'].replace({
    'Debt Consolidation': 'Hợp nhất nợ',
    'Home Improvements': 'Sửa nhà',
    'Business Loan': 'Vay kinh doanh',
    'Buy a Car': 'Mua xe',
    'Buy House': 'Mua nhà',
    'Medical Bills': 'Viện phí',
    'Take a Trip': 'Du lịch',
    'Other': 'Khác',
    'small_business': 'Doanh nghiệp nhỏ',
    'major_purchase': 'Mua sắm lớn',
    'other': 'Khác',
    'wedding': 'Đám cưới',
    'vacation': 'Kỳ nghỉ',
    'Educational Expenses': 'Chi phí giáo dục',
    'moving': 'Chuyển nhà',
    'renewable_energy': 'Năng lượng tái tạo'
})
```
