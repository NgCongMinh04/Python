```python
x = np.array([0, 1, 2, 3])
a1=np.add.reduce(x) # giá trị tổng hợp . 
print(a1)
a2=np.multiply.reduce(x) # nhân tất cả giá trị trong mảng . 
print(a2)
a3=np.multiply.outer([1, 2, 3], [4, 5, 6]) # nhan hai mang 
print(a3)
```

```python
#a1=np.add +
#a2=np.subtract -
#a3=np.multiply *
#A4=np.divide /
#a5=np.floor_divide chia lay phan nguyen
#a6=np.power luy thua 
#a7=np.mod chia lay phan du 
#a8=np.negative phep lay nguoc dau 
# Create sample arrays
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

# Addition (a1 = np.add)
print("Addition:")
print(np.add(arr1, arr2))  # Or simply arr1 + arr2

# Subtraction (a2 = np.subtract)
print("\nSubtraction:")
print(np.subtract(arr1, arr2))  # Or simply arr1 - arr2

# Multiplication (a3 = np.multiply)
print("\nMultiplication:")
print(np.multiply(arr1, arr2))  # Or simply arr1 * arr2

# Division (A4 = np.divide)
print("\nDivision:")
print(np.divide(arr1, arr2))  # Or simply arr1 / arr2 (notice it performs floating-point division)

# Floor division (a5 = np.floor_divide)
print("\nFloor Division (integer quotient):")
print(np.floor_divide(arr1, arr2))

# Power (a6 = np.power)
print("\nPower:")
print(np.power(arr1, 2))  # Or simply arr1 ** 2

# Modulo (a7 = np.mod)
print("\nModulo (remainder):")
print(np.mod(arr1, 2))  # Or simply arr1 % 2

# Negative (a8 = np.negative)
print("\nNegative:")
print(np.negative(arr1))  # Or simply -arr1
```

```python
# nối mảng 
x = np.array([0, 1, 2, 3])
y = np.array([3, 2, 1, 0])
 
print("Phương thức concatenate nhận tham số đầu tiên là một tuple hoặc mảng: ", np.concatenate([x, y]))
print("Phương thức concatenate có thể nối nhiều hơn 2 mảng một lúc: ", np.concatenate([x, y, x, y]))
 
x_grid = np.array([[1,2,3], 
                   [4,5,6]])
print("Sử dụng cho mảng nhiều chiều, nối 2 mảng theo trục đầu tiên (chiều dọc):\n ", np.concatenate([x_grid, x_grid]))
print("Hoặc nối 2 mảng theo trục thứ 2 (chiều ngang): \n", np.concatenate([x_grid, x_grid], axis=1))
x = np.array([1,2,3])
x_grid = np.array([[4,5,6], 
                   [7,8,9]])
 
print("Nối 2 mảng theo chiều dọc:\n ", np.vstack([x, x_grid]))
 
y = np.array([[99],
             [99]])
print("Nối 2 mảng theo chiều ngang: \n", np.hstack([x_grid, y]))
```

```python
x = np.array([[14, 96],
               [46, 82],
               [80, 67],
               [77, 91],
               [99, 87]])
 
print("X = ", x)
 
print("Giá trị lớn nhất: ", np.amax(x))
print("Giá trị bé nhất: ", np.amin(x))
 
print("Giá trị lớn nhất (axis = 0): ", np.amax(x, axis=0))
print("Giá trị lớn nhất (axis = 1): ", np.amax(x, axis=1))
# bo qua phan tu nan 
x = np.array([[14, 96],
               [np.nan, 82],
               [80, 67],
               [77, np.nan],
               [99, 87]])
 
print("X = \n", x)
 
print("Giá trị lớn nhất: ", np.nanmax(x))
print("Giá trị bé nhất: ", np.nanmin(x))
```
